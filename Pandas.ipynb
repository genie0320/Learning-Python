## Series

```python
python_list = [1, 2, 3, 4, 5]

# pandas의 Series는 엑셀의 column과 같은 모양이다.
# 하나의 Series에 들어가는 데이터타입은 모두 같아야 한다.
foo = pd.Series(python_list)
# 0    1
# 1    2
# 2    3
# 3    4
# 4    5
# dtype: int64

foo[2] # 3
```

### Custom index
- Index는 데이터와 길이가 같아야 한다.
- value가 겹쳐도 된다!!! (안되어야 할거 같은데)




```python
bar = ['a', 'b', 'c', 'd', 'd']
pd.Series(python_list, index = ['a', 'b', 'c', 'd', 'd'])
pd.Series(python_list, ['a', 'b', 'c', 'd', 'd'])
pd.Series(python_list, bar)
# a    1
# b    2
# c    3
# d    4
# d    5
# dtype: int64

foo = pd.Series(python_list, bar)
foo['c'] # 3 : 인덱스에 해당하는 값을 돌려준다.
foo['d'] # 여러개의 값이 있는 경우, Series로 돌아온다.
# d    4
# d    5
# dtype: int64
```

```python
# Series from dict : the keys become index.
car = {'key1':'value1', 'key2' : 'value2'}

foo = pd.Series(car)
# key1    value1
# key2    value2
# dtype: object
foo['key1'] # value1
```

#### Conditional Selection
- df == 'Value' : 해당 df의 모든 셀에서 'value' 를 확인하고 bool로 리턴.
- df[df == 'value'] : 해당 df의 모든 셀에서 'value'를 확인하여 value이름으로 리턴.
- df[df == 'value']['cols_name'] : 해당 df의 cols에서 value를 확인하여 value로 리턴.
- df[df == 'value'][['cols_name', 'cols_name2']] : 여러개를 반환.


```python
df == 'Nigeria'
# 	name	phone	email	postalZip	country	currency	numberrange	date
# 0	False	False	False	False	False	False	False	False
# 1	False	False	False	False	False	False	False	False
# 2	False	False	False	False	True	False	False	False

df[df == 'Nigeria']
# 	name	phone	email	postalZip	country	currency	numberrange	date
# 0	NaN	NaN	NaN	NaN	NaN	NaN	NaN	NaN
# 1	NaN	NaN	NaN	NaN	NaN	NaN	NaN	NaN
# 2	NaN	NaN	NaN	NaN	Nigeria	NaN	NaN	NaN

df[df == 'Nigeria']['country']
# 0        NaN
# 1        NaN
# 2    Nigeria
# 3    Nigeria
# 4        NaN
# Name: country, dtype: object

df[df == 'Nigeria'][['country', 'postalZip']]
# 	country	postalZip
# 0	NaN	NaN
# 1	NaN	NaN
# 2	Nigeria	NaN
# 3	Nigeria	NaN
# 4	NaN	NaN
```

#### Multiple Condition
- df[(df['cols'] == 'value') & (df['cols'] == 'value')]
- df[(df['cols'] == 'value') | (df['cols'] == 'value')]

```python
df[df['country'] == 'Sweden']
df[(df['country'] == 'Sweden') | (df['country'] == 'Nigeria')] # or 조건
df[(df['country'] == 'Sweden') & (df['postalZip'] == '223453')] # and 조건

# 매칭하는 데이터의 특정 열값을 리턴.
df[(df['country'] == 'Sweden') & (df['postalZip'] == '223453')]['name']
# 0    Kuame Foster
# Name: name, dtype: object
```
